```python
cities = []

while len(cities)<1000:
    x = (random.choice(range(-90,90))) # randomly generate numbers in range -90 & 90 for latitude
    y = (random.choice(range(-180,180))) # randomly generagte numbers in range -180 & 180 for longitude
    city = citipy.nearest_city(x,y).city_name # look up city to match the lat & long randomly generated
    if city not in cities:
        cities.append(city)
```

```python
lat = []
temp=[]
humidity=[]
windspeed=[]
cloudiness=[]
cityList=[]
cityIdList=[]
i = 0
for cityname in cities:
    response = requests.get("http://api.openweathermap.org/data/2.5/weather?appid=56ee1f766182175e7a690c7c6e9c42b8&units=imperial&q="+cityname).json()
    if (float(response['cod']) <= 200):
        print(cityname, response['id'])
        lat.append(response['coord']['lat'])
        cityList.append(cityname)
        cityIdList.append(response['id'])
        temp.append(response['main']['temp'])
        humidity.append(response['main']['humidity'])
        windspeed.append(response['wind']['speed'])
        cloudiness.append(response['clouds']['all'])
        i += 1
        if (i > 60):
            time.sleep(60)
            i = 0

```

```python
weather_dict = {
    "Latitude": lat,
    "City Name": cityList,
    "Temp": temp,
    "Humidity": humidity,
    "Wind Speed": windspeed,
    "Cloudiness": cloudiness
}
weather_data = pd.DataFrame(weather_dict)
weather_data.head(30)
```

```python
plt.scatter(weather_data["Latitude"], weather_data["Temp"], marker=".")

# Incorporate the other graph properties
plt.title("Temperature(F) vs Latitude")
plt.ylabel("Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Temperature(F) vs,Latitude.png")

# Show plot
plt.show()
```

```python
plt.scatter(weather_data["Latitude"], weather_data["Humidity"], marker=".")
# Incorporate the other graph properties
plt.title("Humidity(%) vs. Latitude")
plt.ylabel("Humidity(%)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Humidity(%) vs. Latitude.png")

# Show plot
plt.show()
```
